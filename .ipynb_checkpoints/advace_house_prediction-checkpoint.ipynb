```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
pd.set_option('display.max_columns', 81)
```

```python
df
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), cmap='viridis')
```

```python
features_with_na = [feature for feature in df.columns if df[feature].isnull().sum()>0]
```

```python
features_with_na
```

```python
numeric_features = [feature for feature in df.columns if df[feature].dtypes !='O']
```

```python
len(numeric_features)
```

```python
date_features = [feature for feature in numeric_features if 'Year' in feature or 'Yr' in feature ]
```

```python
date_features
```

```python
len(df['Alley'].unique())
```

```python
discrete_features = [feature for feature in numeric_features if len(df[feature].unique()) <50]
```

```python
discrete_features
```

```python
len(df['YrSold'].unique())
```

```python
continuous_features = [feature for feature in numeric_features if feature not in discrete_features and feature not in date_features]
```

```python
len(continuous_features)
```

```python
for feature in continuous_features:
    plt.hist(df[feature],bins=25)
    plt.title(feature)
    plt.show()
```

```python
for feature in numeric_features:
    if feature in features_with_na:
        print (feature,':', df[feature].isnull().sum())
```

```python
for feature in numeric_features:
    if feature in features_with_na:
        mean = df[feature].mean()
        print(mean)
        df[feature].fillna(mean, inplace=True)
```

```python
for feature in numeric_features:
    if feature in features_with_na:
        print(df[feature].isnull().sum())
```
