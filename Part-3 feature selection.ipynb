```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('3rd-full-cleaned-data.csv')
```

```python
df.head()
```

```python
df.isnull().sum().sum()
```

```python
df.infer_objects()
```

```python
cat_var = df.select_dtypes(include='object').columns
```

```python
len(cat_var)
```

```python
for i in cat_var:    
    print(len(df[i].unique()))
```

```python
cat_max_10_unique_var = [feature for feature in cat_var if len(df[feature].unique())<10]
```

```python
cat_max_10_unique_var
```

```python
len(cat_max_10_unique_var)
```

```python
dummies_df = pd.DataFrame()
dummies_df = pd.get_dummies(df[cat_max_10_unique_var],drop_first=True)
```

```python
dummies_df.head()
```

```python
cat_more_than_10_var = [feature for feature in cat_var if feature not in cat_max_10_unique_var]
```

```python
cat_more_than_10_var
```

```python
dummies_df2 = pd.DataFrame()
for feature in cat_more_than_10_var:
    labels = df[feature].value_counts().sort_values(ascending=False).head(10).index
    for i in labels:
        dummies_df2[feature+i] = np.where(df[feature]==i,1,0)
```

```python
dummies_df2.head()
```

```python
all_dummies_df = pd.concat([dummies_df, dummies_df2], axis=1)
```

```python
all_dummies_df
```

```python
df1 = pd.concat([df.drop(cat_var, axis=1),all_dummies_df], axis=1)
```

```python
df1.head()
```

```python
df1['SalePrice']
```

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
x = df1.drop('SalePrice',1)
y = df1['SalePrice']
```

```python
feature_selection = SelectKBest(score_func=chi2, k=50)
```

```python
feature_selection.fit(x,y)
```

```python
dir(feature_selection)
```

```python
score = feature_selection.scores_
columns = x.columns
```

```python
feature_analysis = pd.DataFrame((columns, score)).T
```

```python
feature_analysis.sort_values(1, ascending=False).head(80)
```

```python
selected_features = feature_analysis.sort_values(1, ascending=False).head(80)[0]
```

```python
type(selected_features)
```

```python
df2 = df1[selected_features]
df2['SalePrice'] = df1['SalePrice']
```

```python
df2['SalePrice']
```

```python
df2.to_csv('4-datset-after-feature-selection.csv')
```
